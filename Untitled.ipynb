```python
import numpy as np
PI2 = 2 * np.pi
ts = np.linspace(0,10,1000)
def myCHIRP(ts, f0=100, f1=400):
    freqs = np.linspace(f0, f1, len(ts))
    dts = np.diff(ts, prepend=0)
    dphis = PI2 * freqs * dts
    phases = np.cumsum(dphis)
    ys = np.cos(phases)
    return ys
x = myCHIRP(ts)
```

```python
import pandas as pd
pd.Series(x).plot()
```
