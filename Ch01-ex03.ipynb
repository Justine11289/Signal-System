```python
from thinkdsp import CosSignal, SinSignal

cos_sig = CosSignal(freq=400, amp=2.0, offset=0)
sin_sig = SinSignal(freq=800, amp=1.0, offset=0)
```

```python
from thinkdsp import decorate

cos_sig.plot()
decorate(xlabel='Time (s)')
```

```python
sin_sig.plot()
decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix.plot()
decorate(xlabel='Time (s)')
```

```python
mix_wave=mix.make_wave(duration=0.5, start=0, framerate=11025)
mix_wave.make_audio()
```

```python
spectrum = mix_wave.make_spectrum()
spectrum.plot(high=1000)
```

```python
sin_sig += SinSignal(freq=550)
sin_sig.make_wave().make_audio()
```
