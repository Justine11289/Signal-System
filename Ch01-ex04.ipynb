```python
from thinkdsp import read_wave

wave = read_wave('499400__danlucaz__happy-loop-1.wav')
wave.normalize()
wave.make_audio()
```

```python
def stretch(wave, factor):
    wave.ts *= factor
    wave.framerate /= factor
    
stretch(wave, 2.0)
wave.make_audio()
```
